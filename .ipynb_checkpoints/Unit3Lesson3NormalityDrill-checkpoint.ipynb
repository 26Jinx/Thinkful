```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate a random variable with 100 datapoints
gamma_dist = np.random.gamma(1, 1, 100)

# Plot a histogram
plt.hist(gamma_dist)

# Add a vertical line at the mean
plt.axvline(gamma_dist.mean(), color='r', linestyle='solid', linewidth=2)

# Add vertical lines for standard deviation
plt.axvline(gamma_dist.mean()-gamma_dist.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(gamma_dist.mean()+gamma_dist.std(), color='r', linestyle='dashed', linewidth=2)

plt.show()
```

Since this is not a normal distribution, the mean and the standard deviation is skewed toward 0.

```python
# Beta Distribution
beta_dist = np.random.beta(1, 10, 100)
plt.hist(beta_dist)
plt.axvline(beta_dist.mean(), color='purple', linestyle='solid')
plt.axvline(beta_dist.mean()-beta_dist.std(), color='purple', linestyle='dashed')
plt.axvline(beta_dist.mean()+beta_dist.std(), color='purple', linestyle='dashed')
plt.show()
```

```python
# Multinomial Distribution
multinomial_dist = np.random.multinomial(20, [1/12, 4/12, 3/12, 2/12, 1/12, 1/12], 3)
plt.hist(multinomial_dist)
plt.axvline(multinomial_dist.mean(), color='pink', linestyle='dashdot')
plt.axvline(multinomial_dist.mean()-multinomial_dist.std(), color='black', linestyle='dotted')
plt.axvline(multinomial_dist.mean()+multinomial_dist.std(), color='black', linestyle='dotted')
plt.show()
```

```python
# Power distribution
power_dist = np.random.power(5, 10000)
plt.axvline(power_dist.mean(), color='orange', linestyle='dashed')
plt.axvline(power_dist.mean()-power_dist.std(), color='red', linestyle='dotted')
plt.axvline(power_dist.mean()+power_dist.std(), color='red', linestyle='dotted')
plt.hist(power_dist)
plt.show()
```

```python
# Geometric distribution
geo_dist = np.random.geometric(0.33, 10000)
plt.hist(geo_dist)
plt.axvline(geo_dist.mean(), color='green', linestyle='dashdot')
plt.axvline(geo_dist.mean()-geo_dist.std(), color='limegreen', linestyle='dotted')
plt.axvline(geo_dist.mean()+geo_dist.std(), color='limegreen', linestyle='dotted')
plt.show()
```

```python
# Uniform distribution
uni_dist = np.random.uniform(5, 15, 1000)
plt.hist(uni_dist)
plt.axvline(uni_dist.mean(), color='skyblue', linestyle='solid')
plt.axvline(uni_dist.mean()-uni_dist.std(), color='brown', linestyle='dotted')
plt.axvline(uni_dist.mean()+uni_dist.std(), color='brown', linestyle='dotted')
plt.show()
```

```python
# Generate two normally-distributed variables with mean1=5, stddev1=0.5 and mean2=10, stddev2=1
rand1 = np.random.normal(5, 0.5, 1000)
rand2 = np.random.normal(10, 1, 1000)

# Add the two variables to create a third variable
rand3 = rand1 + rand2

# Graph the third varible using a histogram
plt.hist(rand3)

# Plot vertical lines for mean and standard deviation
plt.axvline(rand3.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean()-rand3.std(), color='y', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean()+rand3.std(), color='y', linestyle='dashed', linewidth=2)

# Show plot
plt.show()
```
