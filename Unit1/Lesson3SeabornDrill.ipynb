```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Loading Marvel Comics Wikia Data
customers = pd.read_csv('Mall_Customers.csv')
customers.columns
```

## Choose one variable and plot that variable four different ways

```python
# Set up figure
sns.set(style="ticks")
f, axes = plt.subplots(2, 2, figsize=(15, 15), sharex=False, sharey=False)

x = customers['Age']

# histogram
ax1 = axes[0, 0]
sns.distplot(x, kde=False, color='g', ax=ax1)
ax1.set_title('Fig 1: Histogram')

# bar plot
ax2 = axes[0, 1]
sns.lineplot(data=x, color='r', ax=ax2)
ax2.set_title('Fig 2: Line Plot')

# box plot
ax3 = axes[1, 0]
sns.boxplot(x, color='b', ax=ax3)
ax3.set_title('Fig 3: Box Plot')

# kde plot
ax4 = axes[1, 1]
sns.kdeplot(x, color='purple', ax=ax4)
ax4.set_title('Fig 4: KDE Plot')

plt.suptitle('Age of Subjects', y=0.93, fontsize=16)
```

In order to see the distribution of the surveyed subjects, fig 2 is pretty much useless. While we can see the ages of the individual subject, it is difficult to see any type of trend.

## Choose two continuous variables, and plot them three different ways.

```python
sns.set(style='darkgrid')
f, ax = plt.subplots(figsize=(7, 7))

x = 'Age'
y = 'Spending Score (1-100)'
z = 'Annual Income (k$)'

# Scatterplot
sns.scatterplot(x=x, y=y, size=z, data=customers, color='red', sizes=(1, 150))
ax.set_title('Spending Score and Income by Age', fontsize=15)
```

```python
# Hexbin plot
sns.set(style='ticks')

x = customers['Age']
y = customers['Annual Income (k$)']

g = sns.jointplot(x, y, kind='hex', color='#42f4df', height=7)
g.fig.suptitle('Annual Income vs Age', x=0.47, y=0.8, fontsize=15)
```

```python
# Kernel Density Estimate
sns.set(style='darkgrid')
f, ax = plt.subplots(figsize=(8, 7))

x = customers['Spending Score (1-100)'] 
y = customers['Annual Income (k$)']

cmap = sns.cubehelix_palette(start=2.8, rot=0.1, light=1, as_cmap=True)

g = sns.kdeplot(x, y, cbar=True, cmap=cmap, shade=True)
g.set_title('Annual Income vs Spending Score', fontsize=15)
plt.show()
```

## Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
# Scatter plot with regression line
sns.set(style='dark')

x = 'Annual Income (k$)'
y = 'Spending Score (1-100)'
z = 'Gender'

sns.lmplot(x=x, y=y, hue=z, data=customers)
plt.title('Annual Income vs Spending Score by Gender')
plt.show()
```

```python
# Read new data from suicide.csv
suicides = pd.read_csv('suicide.csv')
suicides.head()
```

```python
# Grouped barplots
sns.set(style='whitegrid')

x = 'year'
y = 'suicides_no'
z = 'sex'

g = sns.catplot(x=x, y=y, hue=z, data=suicides, kind='bar', height=7, aspect=2, palette='muted')
plt.title('Number of Suicides by Year 1985-2016', fontsize=15)
plt.ylabel('Suicides')
plt.show()
```

```python
# Line plot
sns.set_style('darkgrid')
f, ax = plt.subplots(figsize=(15, 7))

sns.lineplot(x=x, y=y, hue=z, data=suicides, ax=ax)
plt.ylabel('Suicides')
plt.title('Number of Suicides by Year 1985-2016', fontsize=15)
plt.show()
```

```python
# Read new data
heart = pd.read_csv('heart.csv')

# Pain type is categorized by number in dataframe
cp_types = {0: 'Typical Angina', 1: 'Atypical Angina', 2: 'Non-anginal Pain', 3: 'Aymptomatic'}

# Replace attempt
heart['cp'] = heart['cp'].replace(cp_types)

# Boxen plot
f, ax = plt.subplots(figsize=(7, 7))
sns.set(style='whitegrid', palette='pastel')

g = sns.boxenplot(x='cp', y='age', data=heart)

# Change legend text
plt.xlabel('Type of Chest Pain', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.title('Types of Chest Pain Reported by Age', fontsize=17)
plt.show()
```

```python
# Grouped boxplots
sns.set(style='white', palette='bright')
f, ax = plt.subplots(figsize=(7, 7))

# Replace 0, 1 with 'Male', 'Female'
heart['sex'] = heart['sex'].replace({0: 'Female', 1: 'Male'})

sns.boxplot(x='cp', y='age', hue='sex', data=heart, palette=['m', 'g'])
plt.title('Types of Chest Pain Reported by Age and Sex', fontsize=17)
plt.xlabel('Types of Chest Pain', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.show()
```

```python
# Violin Plot
sns.set(style='white', palette='pastel')
f, ax = plt.subplots(figsize=(7,7))

sns.violinplot(x='cp', y='age', hue='sex', split=True, data=heart, palette=['pink', 'skyblue'], inner='stick')
sns.despine(offset=10, trim=True)
plt.title('Types of Chest Pain Reported by Age and Sex', fontsize=17)
plt.xlabel('Types of Chest Pain', fontsize=15)
plt.ylabel('Age', fontsize=15)
```

```python
sns.set(style='darkgrid')

x = heart['chol']
y = heart['trestbps']
z = heart['sex']

g = sns.jointplot(x=x, y=y, kind='reg', height=10, color='#ff32e7')
g.plot_joint(sns.scatterplot, hue='sex', data=heart, palette=['#125ba5', '#ff0f3f'], marker='+')
g.ax_joint.collections[0].set_alpha(0) # Hides the plot points created by the first 'reg' plot

plt.suptitle('Resting Blood Pressure vs Cholesterol Level', y=1.03, fontsize='20')
plt.xlabel('Cholesterol Levels (mg/dl)')
plt.ylabel('Resting Blood Pressure (mm/Hg)')
```

This graphic shows that those higher cholesterol levels do tend to have higher resting blood pressure. Statistically, most of the subjects have cholesterol levels of between 200-250mm/dl have a resting blood pressure of around 130mm/Hg.
