```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Loading Marvel Comics Wikia Data
customers = pd.read_csv('Mall_Customers.csv')
customers.columns
```

## Choose one variable and plot that variable four different ways

```python
# Set up figure
sns.set(style="ticks")
f, axes = plt.subplots(2, 2, figsize=(15, 15), sharex=False, sharey=False)

x = customers['Age']

# histogram
ax1 = axes[0, 0]
sns.distplot(x, kde=False, color='g', ax=axes[0, 0])
ax1.set_title('Histogram')

# bar plot
ax2 = axes[0, 1]
sns.lineplot(data=x, color='r', ax=axes[0, 1])
ax2.set_title('Line Plot')

# box plot
ax3 = axes[1, 0]
sns.boxplot(x, color='b', ax=axes[1, 0])
ax3.set_title('Box Plot')

# kde plot
ax4 = axes[1, 1]
sns.kdeplot(x, color='purple', ax=axes[1, 1])
ax4.set_title('KDE Plot')

plt.suptitle('Age of Subjects', y=0.93, fontsize=16)
```
