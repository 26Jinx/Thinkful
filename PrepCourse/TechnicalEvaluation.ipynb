```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('plane_crashes_data.csv')
df
```

```python
# create a new dataframe only including crashes between 1980 to 2009
df1 = df[df['year'] >= 1980]
df1
```

```python
# Get the mean of fatalities from 1980 to 2009
df1['fatalities'].mean()
```

```python
# Group by month
df2 = df.groupby('month')['fatalities'].count()
df2
```

```python
# get the max number of fatalities in a given month
df2.max()
```

```python
# Get the max value of fatalities from a single crash
df['fatalities'].max()
```

```python
df.sort_values('fatalities', ascending=False)
```

```python
# What percentage of crashed flights have had 1 or more fatalities?
df3 = df[df['fatalities'] > 0]
rate = len(df3) / len(df)
print(rate)
```

```python
# Average percent of passengers killed per plane crash for entire period
print((df['fatalities'] / df['aboard']).mean())
```
