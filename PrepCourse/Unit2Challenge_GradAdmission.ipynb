```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Admission_Predict.csv')
df
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(x=df['GRE Score'], y=df['Chance of Admit '])
plt.title('GRE vs Chance of Admission')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admission')

plt.subplot(1, 3, 2)
plt.scatter(x=df['CGPA'], y=df['Chance of Admit '])
plt.title('CGPA vs Chance of Admission')
plt.xlabel('Undergraduate GPA')

plt.subplot(1, 3, 3)
plt.scatter(x=df['LOR '], y=df['Chance of Admit '])
plt.title('Letter of Rec vs Chance of Admit')
plt.xlabel('Strength of LOR')

plt.show()
```

These scatter plots show that GRE scores and the undergraduate GPA is a stronger indicator in predicting chances of getting admitted than letter of recommendation scores. GRE scores higher than ~335 points saw an acceptance rate of about 90%, similar to undergraduate GPA of 9.3 and higher. However, the range of scores given to letters of recommendation, while still higher the better, is not a good indicator of chances of admission. LOR score of 5 had admission chances range of just over 60% to upper 90% while a score of 3 ranged anywhere from 50% to over 95%.

```python
plt.scatter(x=df['SOP'], y=df['TOEFL Score'])
plt.title('TOEFL scores vs Strength of Statement of Purpose')
plt.xlabel('SOP Score')
plt.ylabel('TOEFL Score')
plt.show()
```

TOEFL(Test of English as a Foreign Language) scores compared with the students' Statement of Purpose score did not show a strong relationship. One could theorize that higher TOEFL score will, in general, translate to a higher SOP scores, but the scatter plot shows that this isn't necessarily the case. This could be a result of scoring against an objective(TOEFL) vs a seemingly subjective(SOP assessment by an admissions officer) scale.  

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist([df['SOP'], df['LOR ']], bins=8)
plt.title('SOP and LOR Scores Histogram')
plt.xlabel('Score')

ax = plt.subplot(1, 2, 2)
plt.boxplot([df['SOP'], df['LOR ']])
plt.title('SOP and LOR Scores Box Plot')
plt.ylabel('Score')
ax.set_xticklabels(['SOP', 'LOR'])


plt.show()
```

Examining the histogram and box plots showing the SOP and LOR scores, we can see that both being subjectively scored, most students recieved higher than the numerical average of the score scaling from 1 to 5. The box plots show the median score to be right around 3.5 for both SOP and LOR and the histogram shows that the most frequent score given was above 4.5

```python
plt.boxplot(df['Chance of Admit '])
plt.title('UCLA Graduate School Admission')
plt.ylabel('Chance of Admission Rate')
plt.show()
```

This box plot of all the admission chance rates show that the median rate is around 70% with half of all applicants having a chance of 60-80% getting accepted. Outlier does exist at the bottom of the scale for 2 applicants at 34% chance. Both of these students had all below average scores(GRE, TOEFL, LOR, SOP, CGPA) with one student's LOR score being the only above average score at 4.
